# Let us try to improve the model with pipeline

```python
ohe.fit_transform(df[['Status']])
```

```python
ohe.categories_
```

```python
X.head()
```

```python
from sklearn.compose import make_column_transformer
```

```python
column_trans =  make_column_transformer(
    (OneHotEncoder(), ['Status']),
    remainder = 'passthrough')
```

```python
column_trans.fit_transform(df)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(column_trans, logreg)
```

```python
cross_val_score(pipe, df, y, cv =10, scoring='accuracy').mean()
```

```python
X_new = df.sample(5, random_state=99)
X_new
```

```python
pipe.fit(df, y)
```

```python
pipe.predict(X_new)
```
